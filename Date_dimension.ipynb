```python
import pandas as pd
import numpy as np
```

```python
csv_source_file_path = 'csv_source_files\\'
source_date = pd.read_csv(csv_source_file_path + 'calendar.csv')

source_date["date"] = pd.to_datetime(source_date["date"]).dt.strftime('%Y-%m-%d')
source_date = source_date[ (source_date['date']>'2020-10-31') & (source_date['date']<'2021-03-01') ]
```

```python
source_date.insert(0, "Date_key",  np.arange(len(source_date)))
source_date.head()
```

```python
csv_result_file_path = 'csv_result_files\\'
source_date.to_csv(csv_result_file_path + 'Date_dimension.csv',index=False)
```
