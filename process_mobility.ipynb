```python
import pandas as pd
import numpy as np
```

```python
csv_source_file_path = 'csv_source_files\\'
source_mobility = pd.read_csv(csv_source_file_path + '2020_CA_Region_Mobility_Report.csv')
```

```python
source_mobility.head()
```

```python
#create dimensions
Mobility_dimension = pd.DataFrame(columns=['Metro_area', 'Subregion','Province',     #location
                                           'Retail_and_recreation', 'Grocery_and_pharmacy','Parks', #percent change from baseline
                                           'Transit_stations,','Workplaces','Residential',          #percent change from baseline
                                           'date'])                                                 #saved for later join and need removal
```

```python
#append dimensions
for idx, row in source_mobility.iterrows():
    #Mobility
    mobility_row = []
    
    mobility_row.append(row['metro_area'])
    
    #append only Ontario and Ottawa subregion
    if row['sub_region_2'] is not None:
        if row['sub_region_2'] == 'Toronto Division':
            mobility_row.append('Toronto')
        elif row['sub_region_2'] == 'Ottawa Division':
            mobility_row.append('Ottawa')
        else:
            continue
        mobility_row.append(row['sub_region_1'].strip())  #Province
        mobility_row.append(row['retail_and_recreation_percent_change_from_baseline'])  #percent change from baseline
        mobility_row.append(row['grocery_and_pharmacy_percent_change_from_baseline'])
        mobility_row.append(row['parks_percent_change_from_baseline'])
        mobility_row.append(row['transit_stations_percent_change_from_baseline'])
        mobility_row.append(row['workplaces_percent_change_from_baseline'])
        mobility_row.append(row['residential_percent_change_from_baseline'])
        mobility_row.append(row['date'])
        
        Mobility_dimension.loc[len(Mobility_dimension)] = mobility_row
    
```

```python
Mobility_dimension.insert(0,'Mobility_key',np.arange(len(Mobility_dimension)))
Mobility_dimension.head()
```

```python
csv_result_file_path = 'csv_result_files\\'
Mobility_dimension.to_csv(csv_result_file_path + 'Mobility_dimension.csv',index=False)
```
