```python
import pandas as pd
import numpy as np
```

```python
active = pd.read_csv("active.csv")
```

```python
active.head()
```

```python
Patient_dimension = pd.DataFrame(
    columns=['Patient_key', 'Gender', 'Age_group', 
             'Acquisition_group', 'Outbreak_related'
            ])
```

```python
for idx, row in active.iterrows():
    #Patient
    patient_row = [idx + 100]
    patient_row.append(row['Client_Gender'].strip())
    patient_row.append(row['Age_Group'].strip())
    patient_row.append(row['Case_AcquisitionInfo'].strip())
    patient_row.append(row['Outbreak_Related'])
    Patient_dimension.loc[len(Patient_dimension)] = patient_row
```

```python
Patient_dimension.to_csv("patient_dimension.csv",index=False)
```
