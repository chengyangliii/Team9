```python
import pandas as pd
import numpy as np
```

```python
csv_source_file_path = 'csv_source_files\\'
source_cases = pd.read_csv(csv_source_file_path + '455fd63b-603d-4608-8216-7d8647f43350.csv')
source_cases = source_cases[ (source_cases['Specimen_Date']>'2020-10-31') & (source_cases['Specimen_Date']<'2021-03-01') ]
source_cases = source_cases[ (source_cases['Reporting_PHU_City']=='Toronto') | (source_cases['Reporting_PHU_City']=='Ottawa') ]
```

```python
source_cases.head()
```

```python
source_cases["Outbreak_Related"] = source_cases["Outbreak_Related"].map(lambda x: 1 if (x=="Yes") else 0)

#create dimensions
#Patient_dimension = pd.DataFrame(columns=['id','Gender', 'Age_group','Acquisition_group', 'Outbreak_related'])

source_cases = source_cases[['_id','Client_Gender','Age_Group','Case_AcquisitionInfo','Outbreak_Related']]
source_cases = source_cases.rename(columns={'_id': 'id','Client_Gender':'Gender','Case_AcquisitionInfo':'Acquisition_group','Outbreak_Related':'Outbreak_related'})
```

```python
source_cases.insert(0,'Patient_key',np.arange(len(source_cases)))
source_cases.head()
```

```python
csv_result_file_path = 'csv_result_files\\'
source_cases.to_csv(csv_result_file_path + 'Patient_dimension.csv',index=False)
```
