```python
import pandas as pd
import numpy as np
```

```python
csv_source_file_path = 'csv_source_files/'
source_cases = pd.read_csv(csv_source_file_path + 'positive_cases.csv')
source_cases = source_cases[ (source_cases['Specimen_Date']>'2020-10-31') & (source_cases['Specimen_Date']<'2021-03-01') ]
source_cases = source_cases[ (source_cases['Reporting_PHU_City']=='Toronto') | (source_cases['Reporting_PHU_City']<'Ottawa') ]
```

```python
source_cases.head()
```

```python
#create dimensions
Patient_dimension = pd.DataFrame(columns=['id','Gender', 'Age_group','Acquisition_group', 'Outbreak_related'])
```

```python
source_cases["Outbreak_Related"] = source_cases["Outbreak_Related"].map(lambda x: 'y' if (x=="Yes") else 'n')

Patient_dimension = source_cases[['Row_ID','Client_Gender','Age_Group','Case_AcquisitionInfo','Outbreak_Related']]
```

```python
Patient_dimension = Patient_dimension.rename(columns={'Row_ID': 'id','Client_Gender':'Gender','Case_AcquisitionInfo':'Acquisition_group','Outbreak_Related':'Outbreak_related'})
```

```python
Patient_dimension.insert(0,'Patient_key',np.arange(len(source_cases)))
```

```python
csv_result_file_path = 'csv_result_files/'
source_cases.to_csv(csv_result_file_path + 'Patient_dimension.csv',index=False)
```
