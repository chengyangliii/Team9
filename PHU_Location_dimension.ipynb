```python
import pandas as pd
import numpy as np
```

```python
source_PHU = pd.read_csv("conposcovidloc.csv")
```

```python
source_PHU = source_PHU[ (source_PHU['Specimen_Date']>'2020-10-31') & (source_PHU['Specimen_Date']<'2021-03-01') ]
#source_PHU = source_PHU[(source_PHU["Reporting_PHU_City"] == 'Toronto')]#| (source_PHU["Reporting_PHU_City"] == 'Ottawa')  | (source_PHU["Reporting_PHU_City"] == 'Whitby') | (source_PHU["Reporting_PHU_City"] == 'Newmarket') | (source_PHU["Reporting_PHU_City"] == 'Oakville')|(source_PHU["Reporting_PHU_City"] == 'Mississauga')]
source_PHU = source_PHU[ (source_PHU['Reporting_PHU_City']=='Toronto') | (source_PHU['Reporting_PHU_City']<'Ottawa') ]
```

```python
PHU_Location_dimension = pd.DataFrame(columns=['PHU_Location_key', 
                                               'PHU_name', 'Address', 'City','Postal_Code',
                                               'URL','Latitude','Longitude'])
```

```python
source_PHU = source_PHU[['Reporting_PHU_ID','Reporting_PHU','Reporting_PHU_Address','Reporting_PHU_City','Reporting_PHU_Postal_Code','Reporting_PHU_Website','Reporting_PHU_Latitude','Reporting_PHU_Longitude']]
source_PHU = source_PHU.rename(columns={'Reporting_PHU_ID':'ID','Reporting_PHU' : 'PHU_name','Reporting_PHU_Address' : 'Address','Reporting_PHU_City': 'City','Reporting_PHU_Postal_Code' : 'Postal_Code','Reporting_PHU_Website' : 'URL','Reporting_PHU_Latitude' : 'Latitude','Reporting_PHU_Longitude' : 'Longitude'})
source_PHU.insert(5, "Province", 'Ontario', True)
```
